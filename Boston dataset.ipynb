```python
#importing calculating and implementing libraries
import pandas as pd
import numpy as np
from sklearn import linear_model

#splitting into testing part and training part
from sklearn.model_selection import train_test_split
```

```python
#importing datasets boston house
from sklearn.datasets import load_boston
from sklearn.metrics import mean_squared_error
```

```python
#loading dataset
boston = load_boston()
```

```python
#show keys
boston.keys()
```

```python
#show target
boston.target
```

```python
df_x = pd.DataFrame(boston.data, columns=boston.feature_names)
```

```python
df_y = pd.DataFrame(boston.target)
```

```python
df_x.head(5)
```

```python
df_y[0:6]
```

```python
#repartitioning into training data and test data
x_train, x_test, y_train, y_test = train_test_split(df_x, df_y, test_size=0.2)
```

```python
x_train
```

```python
y_train
```

```python
x_test
```

```python
y_test
```

```python
#linear regression 
regL = linear_model.LinearRegression()
```

```python
#Training our model
regL.fit(x_train, y_train)
```

```python
#calculate predictions with testing data
y_pred=regL.predict(x_test)
```

```python
#show predictions
y_pred
```

```python
#predictions are close to the testing prices 
y_test
```

```python
#calculting the errors
error = np.mean((y_test-y_pred)**2)
```

```python
print(error)
```

```python
mean_squared_error(y_test, y_pred)
```
