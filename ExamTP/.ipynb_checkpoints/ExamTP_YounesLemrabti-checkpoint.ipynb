```python
#1
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn import linear_model, metrics
import seaborn as sns
```

```python
#2
dataset = pd.read_csv('creditcard.csv')
```

```python
#9
dataset.replace({'Class' :{
    0:False,
    1:True
}}, inplace=True)
```

```python
#10
dataset.describe()
```

```python
#11
fraud = np.sum(dataset[['Class']] == True)
fraud
```

```python
nofraud = np.sum(dataset[['Class']] == False)
nofraud
```

```python
perct = (fraud/(fraud+nofraud))*100
perct
```

```python
#12
sns.countplot(dataset[['Class']], dataset[['V2']])
```

```python
#13
from sklearn.preprocessing import StandardScaler
```

```python
norm = StandardScaler()
dataset[['Time']] = norm.fit_transform(dataset[['Time']])
dataset[['Amount']] = norm.fit_transform(dataset[['Amount']])
```

```python
dataset
```

```python
#14
%matplotlib inline
plt.figure(figsize=(50,50))
sns.heatmap(dataset.corr(), annot=True, cmap='Blues')
```

```python
#15
dataset.drop(columns=['V1', 'V3', 'V5', 'V7', 'V6', 'V9', 'V12', 'V14', 'V16', 'V17', 'V18'], inplace=True)
```

```python
dataset.head()
```

```python
#16
df_x = dataset.iloc[:, :-1].values
df_x
```

```python
df_y = dataset.iloc[:, -1].values
df_y
```

```python
#17
x_train, x_test, y_train, y_test = train_test_split(df_x, df_y, test_size=0.5)
```

```python
x_train
```

```python
y_train
```

```python
#18
reg = linear_model.LogisticRegression()
```

```python
reg.fit(x_train, y_train)
```

```python
#19
predict = reg.predict(x_test)
predict
```

```python
#20
metrics.accuracy_score(y_test, predict)
```

```python
#21
mat = metrics.confusion_matrix(y_test, predict)
print(mat)
```

```python
sns.heatmap(mat, annot=True, cmap='Blues')
```

```python
#22
print(metrics.classification_report(y_test, predict))
```
