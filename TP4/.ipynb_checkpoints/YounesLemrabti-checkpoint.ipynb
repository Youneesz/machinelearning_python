```python
x_train, x_test, y_train, y_test = train_test_split(df_x, df_y, test_size=0.2)
```

```python
reg = linear_model.LogisticRegression(max_iter=1000)
```

```python
reg.fit(x_train, y_train)
```

```python
predict = reg.predict(x_test)
```

```python
predict
```

```python
y_test
```

```python
report = metrics.classification_report(y_test, predict)
```

```python
print(report)
```

```python
mat = metrics.confusion_matrix(y_test, predict)
```

```python
metrics.accuracy_score(y_test, predict)
```

```python
plt.figure()
sns.heatmap(mat, annot=True)
```

```python
treeClass = DecisionTreeClassifier()
```

```python
treeClass = treeClass.fit(x_train, y_train)
```

```python
predict = treeClass.predict(x_test)
```

```python
metrics.accuracy_score(y_test, predict)
```

```python
print(metrics.classification_report(y_test, predict))
```

```python
#accuracy score:
#0.81 dans la regression logistique
#0.70 dans l'arbre de decision

#La regression logistique est meilleur en terms de prediction
```

```python
plt.figure(figsize=(20, 20))
tree.plot_tree(treeClass, feature_names=df_x.columns, fontsize=8)
plt.show()
```

```python
df_x1 = dataset.drop(['BloodPressure', 'SkinThickness', 'Outcome'],axis=1)
df_y1 = dataset['Outcome']
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(df_x1, df_y1, test_size=0.2)
```

```python
treeClass = DecisionTreeClassifier(max_depth=5)
treeClass = treeClass.fit(X_train, Y_train)
```

```python
predict = treeClass.predict(X_test)
```

```python
metrics.accuracy_score(Y_test, predict)
```

```python
plt.figure(figsize=(20, 20))
tree.plot_tree(treeClass, feature_names=df_x.columns, fontsize=8)
plt.show()
```
